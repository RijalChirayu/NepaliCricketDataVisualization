```python

# Assuming 'Born_District' column corresponds to district names
players_df['Latitude'] = players_df['Born_District'].map(lambda district: districts_lat_long.get(district, [np.nan, np.nan])[0])
players_df['Longitude'] = players_df['Born_District'].map(lambda district: districts_lat_long.get(district, [np.nan, np.nan])[1])

# Drop rows with NaN values in 'Latitude' and 'Longitude' columns
players_df = players_df.dropna(subset=['Latitude', 'Longitude'])

# Create a folium map centered around Nepal
nepal_map = folium.Map(location=[28.3949, 84.1240], zoom_start=7)

# Function to create player markers
def create_player_marker(row):
    location = [row['Latitude'], row['Longitude']]
    tooltip = f"{row['Name']} - {row['Playing Order']}"

    # Include the image in the popup using the URL
    popup_content = f"<img src='{row['Photo']}' alt='player_image' width='100' height='100'><br>Age: {row['Age']}"

    folium.Marker(location, popup=folium.Popup(popup_content, max_width=300), tooltip=tooltip).add_to(nepal_map)

# Apply the function to each row in the DataFrame
players_df.apply(create_player_marker, axis=1)

# Save and display the map
nepal_map.save('nepal_players_with_district_map.html')
display(nepal_map)
```
